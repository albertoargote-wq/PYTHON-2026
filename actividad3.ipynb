# **8: Descuentos en Tienda de Electrónica**
Enunciado Corto:
Calcular el precio a pagar por productos en una tienda de electrónica aplicando
descuentos según el tipo de producto y, en algunos casos, descuentos adicionales por
productos complementarios.
Desarrollo:
● Mostrar al usuario las opciones de productos:
1. Computadora (con posible descuento adicional en impresora).
2. Televisión (con posible descuento adicional en barra de sonido).
3. Consola de videojuegos (con posible descuento adicional en juego).
● Solicitar al usuario la selección del producto y el precio del producto principal.
● Aplicar el descuento específico para el producto principal.
● Preguntar si se adquiere el producto complementario (impresora, barra de
sonido o juego) y, de ser así, solicitar su precio.
● Aplicar el descuento correspondiente al producto complementario.
● Calcular el total a pagar sumando el precio con descuento del producto
principal y, si aplica, el precio con descuento del complemento.
● Desplegar los precios con descuento y el total final.

```python
print("--- BIENVENIDO A LA TIENDA DE ELECTRÓNICA ---")
print("1. Computadora (Descuento 10%) -> Complemento: Impresora (Descuento 20%)")
print("2. Televisión (Descuento 15%) -> Complemento: Barra de sonido (Descuento 25%)")
print("3. Consola (Descuento 5%) -> Complemento: Juego (Descuento 10%)")
print("---------------------------------------------")

opcion = input("Selecciona el producto principal que deseas comprar (1, 2 o 3): ")

if opcion in ['1', '2', '3']:
    try:
        precio_principal = float(input("Ingresa el precio del producto principal: $"))

        if opcion == '1':
            producto = "Computadora"
            complemento = "Impresora"
            desc_prin = 0.10
            desc_comp = 0.20
        elif opcion == '2':
            producto = "Televisión"
            complemento = "Barra de sonido"
            desc_prin = 0.15
            desc_comp = 0.25
        elif opcion == '3':
            producto = "Consola de videojuegos"
            complemento = "Juego"
            desc_prin = 0.05
            desc_comp = 0.10

        precio_prin_desc = precio_principal - (precio_principal * desc_prin)

        lleva_complemento = input(f"¿Deseas agregar el producto complementario ({complemento})? (s/n): ").lower()

        precio_comp_desc = 0

        if lleva_complemento == 's':
            precio_complemento = float(input(f"Ingresa el precio de {complemento}: $"))
            precio_comp_desc = precio_complemento - (precio_complemento * desc_comp)

        total = precio_prin_desc + precio_comp_desc

        print("\n========= RESUMEN DE COMPRA =========")
        print(f"{producto}: Precio original ${precio_principal:.2f} -> Queda en: ${precio_prin_desc:.2f}")

        if lleva_complemento == 's':
            print(f"{complemento}: Precio original ${precio_complemento:.2f} -> Queda en: ${precio_comp_desc:.2f}")

        print("-" * 37)
        print(f"TOTAL A PAGAR: ${total:.2f}")
        print("=====================================")

    except ValueError:
        print("Error: Ingresaste un valor no válido. Asegúrate de escribir únicamente números para los precios.")

else:
    print("Opción inválida. Por favor, reinicia el programa y elige 1, 2 o 3.")
```

# **9: Descuentos en Tienda de Ropa según Temporada**
Enunciado Corto:
Determinar el precio final de un producto en una tienda de ropa aplicando descuentos
según la temporada y, en ciertos casos, la etiqueta del producto.
Desarrollo:
● Solicitar al usuario el precio original del producto.
● Pedir al usuario que indique la temporada (verano, invierno, primavera u
otoño).
● Según la temporada:
○ En verano, aplicar un descuento fijo a todos los productos.
○ En invierno, solicitar la etiqueta del producto (roja o verde) y aplicar el
descuento correspondiente.
○ En primavera y otoño, solicitar la etiqueta del producto (amarilla) y
aplicar el descuento correspondiente.
● Calcular el precio final aplicando el descuento.
● Desplegar el precio final a pagar.

```python
print("--- TIENDA DE ROPA: DESCUENTOS POR TEMPORADA ---")

try:
    precio_original = float(input("Ingresa el precio original del producto: $"))
    temporada = input("Ingresa la temporada (verano, invierno, primavera, otoño): ").lower()
    descuento = 0.0
    if temporada == "verano":
        descuento = 0.20
    elif temporada == "invierno":
        etiqueta = input("Ingresa el color de la etiqueta (roja o verde): ").lower()
        if etiqueta == "roja":
            descuento = 0.30
        elif etiqueta == "verde":
            descuento = 0.15
        else:
            print("Etiqueta no válida. Se aplicará 0% de descuento.")
    elif temporada == "primavera" or temporada == "otoño":
        etiqueta = input("¿El producto tiene etiqueta amarilla? (si/no): ").lower()
        if etiqueta == "si":
            descuento = 0.10
    else:
        print("Temporada no reconocida. Se aplicará 0% de descuento.")

    precio_final = precio_original - (precio_original * descuento)

    print("\n========= RECIBO =========")
    print(f"Precio original: ${precio_original:.2f}")
    print(f"Descuento aplicado: {descuento * 100:.0f}%")
    print("-" * 26)
    print(f"PRECIO FINAL A PAGAR: ${precio_final:.2f}")
    print("==========================")

except ValueError:
    print("Error: Ingresaste un valor no válido. Por favor, utiliza solo números para el precio.")
```

# **10: Descuentos en Restaurante**

Enunciado Corto:

Calcular el precio final de un menú en un restaurante aplicando descuentos que varían

según el día de la semana y el tipo de menú solicitado.Desarrollo:

● Solicitar al usuario el precio del menú.

● Pedir al usuario que indique el día de la semana en que se realiza la compra.

● Según el día de la semana:

○ Lunes: aplicar 10% de descuento para el menú del día.

○ Martes: aplicar 20% de descuento para el menú infantil.

○ Miércoles: aplicar 15% de descuento para el menú vegetariano.

○ Jueves: aplicar 5% de descuento para el menú del chef.

○ Viernes: aplicar 5% de descuento para el menú del día.

○ Sábados y domingos: no se aplica descuento.

● Calcular el precio final aplicando el descuento correspondiente.

● Desplegar el precio final a pagar.

```python
print("--- RESTAURANTE: DESCUENTOS POR DÍA ---")
try:
    precio_original = float(input("Ingresa el precio del menú: $"))
    dia = input("Ingresa el día de la semana: ").lower()
    tipo_menu = input("Ingresa el tipo de menú (ej. del dia, infantil, vegetariano, del chef, otro): ").lower()
    descuento = 0.0
    if dia == "lunes" and tipo_menu == "del dia":
        descuento = 0.10
    elif dia == "martes" and tipo_menu == "infantil":
        descuento = 0.20
    elif dia == "miércoles" or dia == "miercoles":
        if tipo_menu == "vegetariano":
            descuento = 0.15
    elif dia == "jueves" and tipo_menu == "del chef":
        descuento = 0.05
    elif dia == "viernes" and tipo_menu == "del dia":
        descuento = 0.05
    elif dia in ["sábado", "sabado", "domingo"]:
        descuento = 0.0
    precio_final = precio_original - (precio_original * descuento)
    print("\n========= TICKET DE PAGO =========")
    print(f"Día: {dia.capitalize()} | Menú: {tipo_menu.capitalize()}")
    print(f"Precio original: ${precio_original:.2f}")
    print(f"Descuento aplicado: {descuento * 100:.0f}%")
    print("-" * 34)
    print(f"PRECIO FINAL A PAGAR: ${precio_final:.2f}")
    print("==================================")

except ValueError:
    print("Error: Ingresaste un valor no válido. Por favor, utiliza solo números para el precio.")
```
